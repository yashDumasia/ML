```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/yash/Downloads/customer.csv")
df.head()
```

```python
df = df.iloc[:,2:]
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop("purchased",axis = 1),df["purchased"],test_size = 0.2, random_state = 0 )
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

## Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder 
oe = OrdinalEncoder(categories = [["Poor","Average","Good"],["School","UG","PG"]])

oe.fit(x_train)
x_train_ec = oe.transform(x_train)
x_test_ec = oe.transform(x_test)
```

```python
x_test,x_test_ec
```

```python
oe.categories_
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(y_train)
y_train_ec = le.transform(y_train)
y_test_ec = le.transform(y_test)

# label encoding do on only output column

y_test,y_test_ec
```

```python
le.classes_
```

## 4. One Hot Encoding with top Categories

```python
df["brand"].nunique()
```

```python
count = df["brand"].value_counts()
count
```

```python
z = count[count<100].index
z
# brands which have less than 100 cars
```

```python
w = df["brand"].replace(z,"Other")
# replace the brands which have less than 100 cars with "Other"

c = pd.get_dummies(w) # one hot encoding
c
```
