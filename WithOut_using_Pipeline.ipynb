```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.tree import DecisionTreeClassifier 
```

## With out using Pipeline

```python
df = pd.read_csv("/home/yash/Downloads/tested.csv")
df = df.drop(["PassengerId","Name","Ticket","Cabin"],axis = 1)
df.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop(["Survived"],axis = 1),df["Survived"],test_size = 0.3,random_state = 0)
```

```python
# Pclass, SibSp, Parch 

x_train_PSP = np.array(x_train[["Pclass","SibSp","Parch"]])
x_test_PSP = np.array(x_test[["Pclass","SibSp","Parch"]])
```

```python
# Sex, Embarked

ohe = OneHotEncoder(drop = "first",sparse_output = False)
ohe.fit(x_train[["Sex","Embarked"]])
x_train_SE = ohe.transform(x_train[["Sex","Embarked"]])
x_test_SE = ohe.transform(x_test[["Sex","Embarked"]])
```

```python
# Age, Fare

imputer = SimpleImputer()
imputer.fit(x_train[["Age","Fare"]])
x_train_AF = imputer.transform(x_train[["Age","Fare"]])
x_test_AF = imputer.transform(x_test[["Age","Fare"]]) # fill missing values by that's mean

x_train_AF = pd.DataFrame(x_train_AF)
x_train_AF.columns = ["Age","Fare"] # Add Column Name
x_test_AF = pd.DataFrame(x_test_AF)
x_test_AF.columns = ["Age","Fare"] # Add Column Name

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train[["Age","Fare"]])
x_train_AF = scaler.transform(x_train_AF)
x_test_AF = scaler.transform(x_test_AF)
```

```python
x_train = np.hstack((x_train_PSP,x_train_SE,x_train_AF))
x_train = pd.DataFrame(x_train)

x_test = np.hstack((x_test_PSP,x_test_SE,x_test_AF))
x_test = pd.DataFrame(x_test)
```

```python
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)*100
```
