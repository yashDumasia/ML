## MICE
### (Multivariate Imputation by Chained Equation)

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
```

```python
df = np.round(pd.read_csv("/home/yash/Downloads/50_Startups.csv") [['R&D Spend','Administration','Marketing Spend','Profit']]/10000)
df.head()
```

```python
np.random.seed(9)
df = df.sample(5) 
df = df.iloc[:,0:-1]
df
```

```python
# Remove value to genrate missing value 

df.iloc[1,0] = np.nan
df.iloc[3,1] = np.nan
df.iloc[-1,-1] = np.nan
df.head()
```

```python
# Step 1 - Impute all missing values with mean of respective col

df0 = pd.DataFrame()

df0['R&D Spend'] = df['R&D Spend'].fillna(df['R&D Spend'].mean())
df0['Administration'] = df['Administration'].fillna(df['Administration'].mean())
df0['Marketing Spend'] = df['Marketing Spend'].fillna(df['Marketing Spend'].mean())
```

```python
df0
```

```python
# Remove the col1 imputed value

df1 = df0.copy()

df1.iloc[1,0] = np.nan

df1
```

```python
# Use 4 rows to build a model and use the Except for prediction

x = df1.iloc[[0,2,3,4],1:3]
x
```

```python
y = df1.iloc[[0,2,3,4],0]
y
```

```python
lr = LinearRegression()
lr.fit(x,y)
lr.predict(df1.iloc[1,1:].values.reshape(1,2))
```

```python
# fill missing value in 0th collumn by predicted 
df1.iloc[1,0] = 23.14
df1
```

```python
# Remove the col2 imputed value

df1.iloc[3,1] = np.nan

df1
```

```python
# Use last 3 rows to build a model and use the first for prediction
x = df1.iloc[[0,1,2,4],[0,2]]
x
```

```python
y = df1.iloc[[0,1,2,4],1]
y
```

```python
lr = LinearRegression()
lr.fit(x,y)
lr.predict(df1.iloc[3,[0,2]].values.reshape(1,2))
```

```python
df1.iloc[3,1] = 11.06
```

```python
df1
```

```python
# Remove the col3 imputed value
df1.iloc[4,-1] = np.nan

df1
```

```python
y = df1.iloc[0:4,-1]
y
```
