```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/yash/Downloads/tested.csv")
```

```python
df.sample(5)
```

## Categorical Data :

### Countplot

```python
sns.countplot(x=df["Survived"])
df["Survived"].value_counts()
```

### Pie Chart

```python
df["Pclass"].value_counts().plot(kind = "pie",autopct = "%.2f")
# autopct print % of value in pie chart and 2 means there are 2 decimal places
```

## Numerical Data :

### Histogram

```python
plt.hist(df["Age"],bins = 15) # bins = 15 means there are 50 bars in histogram
```

### Displot

```python
sns.displot(df["Age"], kde = True) # kde = True means it make curve on displot 
# Curve is called PDF (Probability Density Function)
plt.grid(color = "black", linestyle = "--", linewidth = 0.5,alpha = 0.5) 
```

### Bowplot

```python
sns.boxplot(x = df["Age"])
plt.title("BoxPlot")
```
