## Working on Date

```python
# 1. Extract Year

date["year"] = date["date"].dt.year
date.head()
```

```python
# 2. Extract Month

date["month"] = date["date"].dt.month
date.head()
```

```python
date["month_name"] = date["date"].dt.month_name()
date.head()
```

```python
# 3. Extract Day

date["day"] = date["date"].dt.day
date.head()
```

```python
date["day_of_week"] = date["date"].dt.day_of_week
date.head()
```

```python
date["name_of_day"] = date["date"].dt.day_name()
date.head()
```

```python
date["day_of year"] = date["date"].dt.day_of_year
date.head()
```

```python
# is weekend...??

date["weekend..??"] = np.where(date["name_of_day"].isin(["Saturday","Sunday"]),"Yes","No")
date.head()
```

```python
# 4. Extract Week of the Year

date["week_of_year"] = date["date"].dt.isocalendar().week
date.head()
```

```python
# 5. Extract Quarter

date["quarter"] = date["date"].dt.quarter
# A quarter is simply a way to divide a full year into four parts, each part being 3 months.
date.drop(columns=["product_id","city_id","orders"]).head()
```

```python
# Semister 1 or 2..??

date["semister"] = np.where(date["quarter"].isin([1,2]),1,2)
date.drop(columns=["product_id","city_id","orders"]).head()
```
