## Handle Mix Data

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/home/yash/Downloads/Titanic_Mixed_Data.csv")
```

```python
df = df[["Cabin","Ticket","Number","Survived"]]
df.tail()
```

```python
df["Number"].unique()
```

```python
fig = df["Number"].value_counts().plot.bar()
```

```python
# Extract Numerical Prat:-
df["Number_Numerical"] = pd.to_numeric(df["Number"],errors = "coerce",downcast = "integer")
# pd.to_numeric(arg, errors='raise', downcast=None) :-
#     ->arg → the data you want to convert (like df["column_name"])
#     ->errors → what to do if something can’t be converted
#         -'raise' → (default) gives an error
#         -'coerce' → replaces non-numeric values with NaN
#         -'ignore' → leaves the data unchanged
#     ->downcast → tries to make data smaller (like turning int64 → int8 or int16(to save memory))
df.tail()
```

```python
# Extract Categorical Prat:-
df["Number_Categorical"] = np.where(df["Number_Numerical"].isnull(),df["Number"],np.nan)
# np.where(condition, value_if_true, value_if_false)
df.tail()
```

```python
df["Cabin"].unique()
```

```python
df["Ticket"].unique()
```

```python
df["Cabin_cat"] = df["Cabin"].str[0]
# .str[0] → takes the first character from each value
df["Cabin_num"] = df["Cabin"].str[1:3]
# .str[1:3] → slices characters from index 1 up to (but not including) index 3 of each string
df.tail()
```

```python
print(df["Cabin_cat"].unique())
print(df["Cabin_cat"].value_counts())
df["Cabin_cat"].value_counts().plot(kind = "bar")
```

```python
df["Ticket_num"] = df["Ticket"].apply(lambda s: s.split()[-1])
df["Ticket_num"] = pd.to_numeric(df["Ticket_num"],errors = "coerce",downcast = "integer")

df["Ticket_cat"] = df["Ticket"].apply(lambda s: s.split()[0])
df["Ticket_cat"] = np.where(df["Ticket_cat"].str.isdigit(),np.nan,df["Ticket_cat"])

df.tail()
```
