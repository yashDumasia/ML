```python
import numpy as np
import pandas as pd
import seaborn as sns 

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("/home/yash/Downloads/Titanic.csv")[["Age","Pclass","SibSp","Parch","Survived"]]
df.sample(5,random_state = 1)
```

```python
df.dropna(inplace = True)
```

```python
x = df.drop(columns = "Survived")
y = df["Survived"]
```

```python
print(np.mean(cross_val_score(LogisticRegression(),x,y,scoring = "accuracy",cv = 20))*100)
```

## By Using Feature Construction

```python
new_df = df.copy()
new_df["Family"] = np.where(((df["SibSp"] == 0) & (df["Parch"] == 0)),
                        0,
                        np.where(((0 < df["SibSp"] + df["Parch"]) & (df["SibSp"] + df["Parch"] <= 3)),
                                 1,
                                 2))
```

```python
new_df.sample(7,random_state = 1)
```

```python
new_df.drop(columns = ["SibSp","Parch"],inplace = True)
```

```python
X = new_df.drop(columns = "Survived")
Y = new_df["Survived"]
```

```python
print(np.mean(cross_val_score(LogisticRegression(),X,Y,scoring = "accuracy",cv = 20))*100)
```
