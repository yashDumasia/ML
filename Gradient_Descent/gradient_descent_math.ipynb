```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples = 4, n_features = 1, n_informative = 1, n_targets = 1,noise = 80,random_state = 13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# OLS (Oridinary Least Square)
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red')
```

```python
# Gradient Descent 
# Assume slop is constant m = 78.35
# And let's assume the starting value for intercept b = 0
b = 0
y_pred1 = ((78.35 * x) + b).reshape(4)
```

```python
y_pred1
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = "Red")
plt.plot(x,y_pred1,color = "Blue")
```

```python
m = 78.35

loss_slop = -2 * np.sum(y-m*x.ravel() - b)
loss_slop
```

```python
# let's take learning rate = 0.1
lr = 0.1

step_size = loss_slop * lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = m * x + b
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = "Red")
plt.plot(x,y_pred1,color = "Blue")
plt.plot(x,y_pred2,color = "Green")
```

```python
# Now apply loop to do this thing 
b = 0
m = 78.35
lr = 0.1

for i in range(0,10):
    loss_slop = -2 * np.sum(y-m*x.ravel() - b)
    step_size = loss_slop * lr
    b = b - step_size
    y_pred_loop = m * x + b
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = "Blue")
plt.plot(x,y_pred_loop,color = "Red")
```
